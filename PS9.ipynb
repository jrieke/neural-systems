```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### 1. a)

```python
t = np.linspace(0, 7, 1000)

m = 0.96 * (1 - np.exp(-t / 0.25))
h = 0.6 * np.exp(-t / 0.84)
n = 0.89 - 0.38 * np.exp(-t / 1.7)

plt.plot(t, m, label='m')
plt.plot(t, h, label='h')
plt.plot(t, n, label='n')
plt.xlabel('t / ms')
plt.ylabel('x')
plt.legend(loc='center right')
```

#### b)

```python
g_Na_bar = 70.7
g_K_bar = 24.3

g_Na = g_Na_bar * m**3 * h
g_K = g_K_bar *n**4

plt.plot(t, g_Na, label='Na')
plt.plot(t, g_K, label='K')
plt.xlabel('t / ms')
plt.ylabel('$g_x$ / (mS / cm^2)')
plt.legend(loc='center right')
```

#### c)

As can be seen in the plot, the steady state values are g_Na = 15 mS / cm^2 and g_K = 0.

#### d)

g_Na reaches its maximum after approximately 0.75 ms with a value of 13 mS / cm^2.
